# Analyse

- ## Les features extraites avec BERT et USE permettent d'entrainer un classificateur qui convergent vers les mêmes résultats (precision , recall, f1).
- ## Les features BERT paraissent plus "riches" : avec 9 features extraites par PCA on arrive à des résultats quasi identiques qu'avec 38 features pour USE.
- ## L'ajout du vader score n'améliore pas les performances passés un certain seuil, l'information stockée doit donc être redondante par rapport à certaines features.
- ## La régression logistique donne les meilleurs résultats (performances et coût en ressources).
